```python
import matplotlib.pyplot as plt
from random import choice
%matplotlib inline
```

```python
def transform_1(point):
    return [0.5*point[0] , 0.5*point[1]]

def transform_2(point):
    return [0.5*point[0] + 0.5 , 0.5*point[1] + 0.5]

def transform_3(point):
    return [0.5*point[0] + 1 , 0.5*point[1]]
```

```python
point = [0,0]

transforms = [transform_1,transform_2,transform_3]

a = [0]
b = [0]

for i in range(500_000):
    point = choice(transforms)(point)
    a.append(point[0])
    b.append(point[1])

plt.rc('figure',figsize = (10,10))
plt.plot(a,b,'o')
plt.show()
```
