```python
def random_walk(steps):
    avail = [1,-1]
    walk = [choice(avail)]
    for i in range(1,steps):
        current = walk[i-1] + choice(avail)
        walk.append(current)
    plt.title("One Random Walk of {} steps".format(steps))
    plt.plot(walk)
    plt.show()


def many_random_walks(number,steps):
    choices = [1,-1]
    walk = []
    walk.append(choice(choices))
    
    for j in range(number):
        for i in range(1,steps):
            current = walk[i-1] + choice(choices)
            walk.append(current)
        plt.plot(walk,label = f'Walk {j+1}')
        walk = [choice(choices)]
    
    plt.rc('figure',figsize = (20,10))
    plt.title("{} Random Walks of {} steps".format(number,steps))
    #plt.legend(loc='lower left')
    plt.xlabel("Number of steps")
    plt.ylabel("Distance from Origin")
    plt.show()

many_random_walks(500,1000)
```
